## 1. Data Exploration

### 1.1 Importing Data-Related Python Libraries


Python contains specific libraries that are particularly helpful for data analysis .
Think of a library as a pre-packaged set of tools that are readily available to use without having to build them from scratch--similar to boxed cake mixes.

````python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

Most datsets will be saved as .xlsx or .csv file. Fortunately, Python allows us to import data from different formats. We will use **Pandas** to help us with this task. 

#### Importing the candy_data_kaggle.csv file 

This will load the dataset and convert it into a dataframe, which will be saved in the *candy_data* variable.

````python
candy_data = pd.read_csv("https://raw.githubusercontent.com/achavez8/camp10.0/main/candy_data_kaggle.csv")
````

## 2. Useful Plots
The libraries we imported (namely Matplotlib and Seaborn) can help us create amazing visualizations.

####The Correlation Map 



````python
plt.figure(figsize=(15,11))
plt.suptitle("Examining the correlation between variables", fontsize=15)
sns.heatmap(candy_data.corr(), annot=True, fmt="0.2f", cmap="PuOr", vmin=-1.0, vmax=1.0)
plt.tight_layout(pad=5.0)
plt.show()
````

#### Vertical Histogram

````python
plt.figure(figsize=(8,15))
sns.barplot(x=candy_data["winpercent"],y=candy_data["competitorname"],order=candy_data[["competitorname","winpercent"]].sort_values(by="winpercent",ascending=False).iloc[:,0],palette="PRGn")
plt.xlabel("Win percentage");
plt.ylabel("");
plt.title("Candy Ranking");
````
